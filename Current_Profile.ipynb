```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# taken form the "electron_30MeV_10cm.root" event number 39

reset_30_10 = np.genfromtxt("datafiles/currentprofile/electron_30MeV_10cm_reset.csv",delimiter=',')
curre_30_10 = np.genfromtxt("datafiles/currentprofile/electron_30MeV_10cm_current.csv",delimiter=',')

reset_30_150 = np.genfromtxt("datafiles/currentprofile/electron_30MeV_150cm_reset.csv",delimiter=',')
curre_30_150 = np.genfromtxt("datafiles/currentprofile/electron_30MeV_150cm_current.csv",delimiter=',')

ElectronCharge_ = 1.60217662e-19
```

```python
fig = plt.figure(figsize=(18,10))

ax1 = fig.add_subplot(221)
ax1.plot(curre_30_10[:,0]*1e6,curre_30_10[:,1], color='k', lw=3, label='Current profile')

for reset in reset_30_10[:,1]:
    ax1.axvline(reset*1e6, color='darkred', lw=2, ls='-',alpha=0.4)

ax1.set_ylabel("Current [nA]", fontsize=18)
ax1.tick_params('y',labelsize=16)
ax1.tick_params('x',labelsize=0)
ax1.grid()

ax2 = fig.add_subplot(223, sharex=ax1)
ax2.plot(curre_30_10[:,0]*1e6,curre_30_10[:,2], color='k', lw=3, label='Current profile')
ax2.step(reset_30_10[:,1]*1e6, reset_30_10[:,0]*6250*ElectronCharge_/10e-9*1e9,  
         where='post', color='darkred',lw=3, label="C++ RTD")

# ax2.legend(loc='upper right', fontsize=18)
ax2.set_ylabel("Cumulative current [nA]",fontsize=18)
ax2.set_xlabel(r"Time [$\mu$s]", fontsize=18)
ax2.tick_params(labelsize=16)
ax2.grid()
ax1.set_xlim(37,62)

##############################################################################
##############################################################################
##############################################################################
ax3 = fig.add_subplot(222, sharey=ax1)
ax3.plot(curre_30_150[:,0]*1e6,curre_30_150[:,1], color='k', lw=3, label='Current profile')

for reset in reset_30_150[:,1]:
    ax3.axvline(reset*1e6, color='darkblue', lw=2, ls='-',alpha=0.4)

# ax3.set_ylabel("Current [nA]", fontsize=18)
ax3.tick_params('y',labelsize=16)
ax3.tick_params('y',labelsize=0)
ax3.tick_params('x',labelsize=0)
ax3.grid()

ax4 = fig.add_subplot(224, sharex=ax3, sharey=ax2)
ax4.plot(curre_30_150[:,0]*1e6,curre_30_150[:,2], color='k', lw=3, label='Current profile')
ax4.step(reset_30_150[:,1]*1e6, reset_30_150[:,0]*6250*ElectronCharge_/10e-9*1e9,  
         where='post', color='darkblue',lw=3, label="C++ RTD")

# ax2.legend(loc='upper right', fontsize=18)
# ax4.set_ylabel("Cumulative current [nA]",fontsize=18)
ax4.set_xlabel(r"Time [$\mu$s]", fontsize=18)
ax4.tick_params('y',labelsize=0)
ax4.tick_params('x',labelsize=16)
# ax4.tick_params(labelsize=16)
ax4.grid()
ax3.set_xlim(885,914)


fig.tight_layout()
# plt.savefig("plots/30MeV_electron_currentprofile.png", dpi=400, bbox_inches='tight')
# plt.savefig("plots/30MeV_electron_currentprofile.pdf", bbox_inches='tight')
plt.show()
```
