```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
datarate = np.load("datafiles/Datarate.npy")
```

```python
cuts = np.arange(50)[1:]

fig = plt.figure(figsize=(8,7))
ax1 = fig.add_subplot(111)
ax1.scatter(cuts,datarate[1:], color='darkorchid')

ax1.semilogy()

ax1.set_ylabel("Resets / APA / 10sec",fontsize=18)
ax1.tick_params(labelsize=18)
ax1.set_xlabel("Reset cut threshold",fontsize=18)

x_major_ticks = np.arange(0, 51,10)
ax1.set_xticks(x_major_ticks)
ax1.set_xticks(np.arange(0, 51,5), minor=True)

y_major_ticks = np.array([0.01, 0.1, 1, 10, 100, 1000, 10000, 100000])
ax1.set_yticks(y_major_ticks)


ax1.set_xlim(0,49.99)
# ax1.set_ylim(1e-2,1e5)
ax1.set_ylim(3e-2,1e5)
ax1.grid(which='both',linestyle=':')



float_formatter = "{:.5f}".format
axDual = ax1.twinx()
axDual.semilogy()
# 64 bits in word, 8 bytes in 64 bits, 1000 to make it kb, 10 for 10 seconds.
newpos = y_major_ticks*64/8/1000/10
# newpos = [float_formatter(pos) for pos in newpos]
axDual.set_yticks(y_major_ticks)
axDual.set_yticklabels(newpos)
axDual.tick_params(axis='y',labelsize=18)
axDual.yaxis.set_ticks_position('right') # set the position of the second x-axis to bottom
axDual.yaxis.set_label_position('right') # set the position of the second x-axis to bottom
axDual.spines['right'].set_position(('outward', 0))
axDual.set_ylabel('Data rate / APA [kBps]',fontsize=18)
axDual.set_ylim(ax1.get_ylim())
# axDual.set_yticklabels(newpos)

# plt.savefig("plots/QPIX_datarate.png",bbox_inches='tight',dpi=400)
# plt.savefig("plots/QPIX_datarate.pdf",bbox_inches='tight')
plt.show()
```
