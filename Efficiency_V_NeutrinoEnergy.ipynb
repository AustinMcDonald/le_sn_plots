```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
DUNE = np.genfromtxt("datafiles/DUNE_Efficiency_V_NeutrinoEnergy.csv", skip_header=2, delimiter=',')
QPIX = np.load("datafiles/QPIX_Efficiency_V_NeutrinoEnergy.npy")

Mybins = np.arange(0,61,0.5)
bin_c = (Mybins[:-1] + Mybins[1:])/2
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)

ax.errorbar(bin_c, QPIX[1],yerr=0,fmt="s",color='r',label="5MeV reco",alpha=0.7,ms=8,ls='--')

ax.errorbar(DUNE[:,0],DUNE[:,1],fmt="<", label="DUNE 5MeV")

ax.errorbar(bin_c, QPIX[0],yerr=0,fmt="d",color='k',label="1MeV reco",alpha=0.7,ms=8,ls='--')

ax.errorbar(DUNE[:,2],DUNE[:,3],fmt=">", label="DUNE 1MeV")



ax.legend(loc='lower right',fontsize=18)
ax.set_ylim(0,1.1)
# ax.xlim(-2,60)
ax.set_xlim(-1,30)
ax.tick_params(axis='both',which='both',labelsize=18, length=4, width=1.5)
ax.set_xticks(np.arange(0,30),minor=True)
ax.set_yticks(np.arange(0,1.1, 0.05),minor=True)

ax.grid()
ax.set_xlabel("Neutrino Energy [MeV]",fontsize=18)
ax.set_ylabel("Efficiency",fontsize=18)


plt.tight_layout()
# plt.savefig("plots/QPIX_Efficiency_V_NeutrinoEnergy.png",dpi=400,bbox_inches='tight')
# plt.savefig("plots/QPIX_Efficiency_V_NeutrinoEnergy.pdf",bbox_inches='tight')
plt.show()
```
