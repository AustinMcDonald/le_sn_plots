```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Wvalue = 23.6
Reset = 6250

time_slice = np.arange(0,25e-6, 0.25e-6)[1:]
reset_threshold_array = np.arange(2,25, 1)

# F is the full, C is the low energy
Eff_F, Pur_F, Eff_C, Pur_C = np.load("datafiles/QPIX_Event_EfficiencyPurity.npy")
```

```python
n = 18
colors = plt.cm.jet(np.linspace(0,1,n))

n = 15
colors = plt.cm.magma(np.linspace(0,1,n))


fig5 = plt.figure(constrained_layout=False,figsize=(28,8))
widths  = [1,1]
heights = [1,1]
spec5 = fig5.add_gridspec(ncols=2, nrows=2, width_ratios=widths, height_ratios=heights,
                         left=0.05, right=0.48, wspace=0.19, hspace=0.1)

ax1 = fig5.add_subplot(spec5[0, 0])
ax2 = fig5.add_subplot(spec5[0, 1],sharey=ax1)
ax3 = fig5.add_subplot(spec5[1, 0],sharex=ax1)
ax4 = fig5.add_subplot(spec5[1, 1],sharey=ax3,sharex=ax2)

ax1.plot(time_slice*1e6, Eff_F[:,11], label="13 resets", color=colors[0], lw=2)
ax1.plot(time_slice*1e6, Eff_F[:,10], label="12 resets", color=colors[3], lw=2)
ax1.plot(time_slice*1e6, Eff_F[:,9],  label="11 resets", color=colors[6], lw=2)
ax1.plot(time_slice*1e6, Eff_F[:,7],  label="9 resets" , color=colors[9], lw=2)
ax1.plot(time_slice*1e6, Eff_F[:,4],  label="6 resets" , color=colors[12], lw=2)
# ax1.plot(time_slice*1e6, Eff_F[:,1],  label="3 resets" , color=colors[12], lw=2)

ax2.plot(time_slice*1e6, Pur_F[:,11], label=" 13 resets \n[1.92 MeV]", color=colors[0], lw=2)
ax2.plot(time_slice*1e6, Pur_F[:,10], label=" 12 resets \n[1.77 MeV]", color=colors[3], lw=2)
ax2.plot(time_slice*1e6, Pur_F[:,9],  label=" 11 resets \n[1.62 MeV]", color=colors[6], lw=2)
ax2.plot(time_slice*1e6, Pur_F[:,7],  label="  9 resets \n[1.33 MeV]" , color=colors[9], lw=2)
ax2.plot(time_slice*1e6, Pur_F[:,4],  label="  6 resets \n[0.89 MeV]" , color=colors[12], lw=2)
# ax2.plot(time_slice*1e6, Pur_F[:,1],  label="  3 resets \n[0.44 MeV]" , color=colors[12], lw=2)

ax3.plot(time_slice*1e6, Eff_C[:,11], label="13 resets", color=colors[0], lw=2)
ax3.plot(time_slice*1e6, Eff_C[:,10], label="12 resets", color=colors[3], lw=2)
ax3.plot(time_slice*1e6, Eff_C[:,9],  label="11 resets", color=colors[6], lw=2)
ax3.plot(time_slice*1e6, Eff_C[:,7],  label="9 resets" , color=colors[9], lw=2)
ax3.plot(time_slice*1e6, Eff_C[:,4],  label="6 resets" , color=colors[12], lw=2)
# ax3.plot(time_slice*1e6, Eff_C[:,1],  label="3 resets" , color=colors[12], lw=2)

ax4.plot(time_slice*1e6, Pur_C[:,11], label="13 resets", color=colors[0], lw=2)
ax4.plot(time_slice*1e6, Pur_C[:,10], label="12 resets", color=colors[3], lw=2)
ax4.plot(time_slice*1e6, Pur_C[:,9],  label="11 resets", color=colors[6], lw=2)
ax4.plot(time_slice*1e6, Pur_C[:,7],  label="9 resets" , color=colors[9], lw=2)
ax4.plot(time_slice*1e6, Pur_C[:,4],  label="6 resets" , color=colors[12], lw=2)
# ax4.plot(time_slice*1e6, Pur_C[:,1],  label="3 resets" , color=colors[12], lw=2)


ax2.legend(loc='upper left', bbox_to_anchor=(-1.31, 1.35), fontsize=18, ncol=5)

# ax1.set_yticks(np.arange(0,1.01,.1))
ax1.set_yticks(np.arange(0,1.01,.1),minor=True)
ax2.set_yticks(np.arange(0,1.01,.1),minor=True)
ax3.set_yticks(np.arange(0,1.01,.1),minor=True)
ax4.set_yticks(np.arange(0,1.01,.1),minor=True)

ax1.tick_params(axis='y',which='both',labelsize=15)
ax1.tick_params(axis='x',labelsize=0)
ax2.tick_params(axis='y',which='both',labelsize=15)
ax2.tick_params(axis='x',labelsize=0)
ax3.tick_params(axis='y',labelsize=15)
ax3.tick_params(axis='x',labelsize=15)
ax4.tick_params(axis='y',which='both',labelsize=15)
ax4.tick_params(axis='x',labelsize=15)

ax1.set_ylabel(r"Event I.D. efficiency", fontsize=18)
ax2.set_ylabel(r"Event I.D. purity", fontsize=18)
ax3.set_ylabel(r"Event I.D. efficiency", fontsize=18)
ax4.set_ylabel(r"Event I.D. purity", fontsize=18)

ax3.set_xlabel(r"Time window [$\mu$s]", fontsize=18)
ax4.set_xlabel(r"Time window [$\mu$s]", fontsize=18)

ax1.set_ylim(0,1.05)
ax2.set_ylim(0,1.05)
ax3.set_ylim(0,1.05)
ax4.set_ylim(0,1.05)

ax1.set_xlim(0,19.99)
ax2.set_xlim(0,19.99)
# ax2.set_xlim(1201/10.0,1300/10.0)

ax1.grid(which='both',linestyle=':')
ax2.grid(which='both',linestyle=':')
ax3.grid(which='both',linestyle=':')
ax4.grid(which='both',linestyle=':')

plt.tight_layout()
# plt.savefig("plots/Event_EfficiencyPurity_bubble.png",bbox_inches='tight',dpi=400)
# plt.savefig("plots/Event_EfficiencyPurity_bubble.pdf",bbox_inches='tight')
plt.show()
```

```python
fig = plt.figure(figsize=(15,7))

ax1 = fig.add_subplot(111)
ax2 = ax1.twinx()

ax1.plot(reset_threshold_array, Eff_F[11], label="Full Spectrum Event ID efficiency", lw=3, color='k', ls='-')
ax1.plot(reset_threshold_array, Eff_C[11], label="Low Energy Event ID efficiency", lw=3, color='k', ls='--')


ax2.plot(reset_threshold_array, Pur_F[11], label="Full Spectrum Event ID purity", lw=3, color='darkred', ls='-')
ax2.plot(reset_threshold_array, Pur_C[11], label="Low Energy Event ID purity", lw=3, color='darkred', ls='--')

# ax3.set_yticks(np.arange(0,1.01,.1),minor=True)

ax1.set_ylabel("Event I.D. efficiency",fontsize=18)
ax1.tick_params(labelsize=18)
ax2.set_ylabel("Event I.D. purity",fontsize=18, color="darkred")
ax2.tick_params(labelsize=18, labelcolor='darkred')
ax1.set_xlabel("Cluster reset threshold",fontsize=18)

x_major_ticks = np.arange(2, 25,2)

ax1.set_yticks(np.arange(0,1.01,.1))
ax2.set_yticks(np.arange(0,1.01,.1))

ax1.set_xticks(x_major_ticks)
ax1.set_xticks(np.arange(2,24,1),minor=True)

ax1.set_xlim(2,24)
ax1.set_ylim(0,1.05)
ax2.set_ylim(0,1.05)
ax1.grid(which='both',linestyle=':')

# ax1.legend(loc='center left', fontsize=18)

lines_labels = [ax.get_legend_handles_labels() for ax in fig.axes]
lines, labels = [sum(lol, []) for lol in zip(*lines_labels)]
plt.legend( lines, labels, loc = 'lower right', fontsize=18)

float_formatter = "{:.3f}".format
axDual = ax1.twiny()
newpos = x_major_ticks*Wvalue*Reset/1e6
newpos = [float_formatter(pos) for pos in newpos]
axDual.set_xticks(x_major_ticks)
axDual.set_xticklabels(newpos)
axDual.tick_params(axis='x',labelsize=18)
axDual.xaxis.set_ticks_position('bottom') # set the position of the second x-axis to bottom
axDual.xaxis.set_label_position('bottom') # set the position of the second x-axis to bottom
axDual.spines['bottom'].set_position(('outward', 55))
axDual.set_xlabel('Equivalent energy [MeV]',fontsize=18)
axDual.set_xlim(ax1.get_xlim())

plt.tight_layout()
# plt.savefig("plots/Event_EfficiencyPurity_reset.png",bbox_inches='tight',dpi=400)
# plt.savefig("plots/Event_EfficiencyPurity_reset.pdf",bbox_inches='tight')
plt.show()
```
