```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
all_truth, all_resets, all_corr = np.load("datafiles/QPIX_2D_Energy_Reco.npy")
```

```python
x_bins = np.arange(0, 100, 0.5)
y_bins = np.arange(0, 100, 0.5)
# cmap = plt.get_cmap('cividis')
cmap = plt.get_cmap('afmhot')

fig = plt.figure(figsize=(16,7))
ax1 = fig.add_subplot(121)
# ax1.plot(Etru_150, Erec_150, lw=2, label='Uncorrected single electrons')
ax1.plot(x_bins, x_bins, lw=2)

counts, x_edges, y_edges = np.histogram2d(all_truth, all_resets, bins=(x_bins, y_bins))
x, y = np.meshgrid(x_bins, y_bins)
h1 = ax1.pcolormesh(x, y, np.ma.masked_where(counts == 0, counts).T, shading='auto', cmap=cmap)  # with mask

cbar1 = plt.colorbar(h1)
cbar1.ax.tick_params(labelsize=22)
cbar1.set_label(label='Counts per 500keV', fontsize=22)

# ax1.legend(loc='upper right',fontsize=18)
ax1.set_ylabel(r'Reconstructed Energy [MeV]',fontsize=22)
ax1.set_xlabel(r'Neutrino Energy [MeV]',fontsize=22)
ax1.tick_params(labelsize=22)
# ax1.legend(loc="upper left", fontsize=18)




ax2 = fig.add_subplot(122)
# correction = Etru_150/(my_curve(Etru_150, *popt_150))
# ax2.plot(Etru_150, Erec_150*correction, lw=2, label='Corrected single electrons')
ax2.plot(x_bins, x_bins, lw=2)

counts, x_edges, y_edges = np.histogram2d(all_truth, all_corr, bins=(x_bins, y_bins))
x, y = np.meshgrid(x_bins, y_bins)
h2 = ax2.pcolormesh(x, y, np.ma.masked_where(counts == 0, counts).T, shading='auto', cmap=cmap)  # with mask

#h2 = ax2.hist2d(all_truth, all_corr, bins=bins, cmap=cmap)
cbar2 = plt.colorbar(h2)
cbar2.set_label(label='Counts per 500keV', fontsize=22)
cbar2.ax.tick_params(labelsize=22)

# ax2.legend(loc='upper right',fontsize=18)
# ax2.set_ylabel(r"Counts [arb.]",fontsize=18)
ax2.set_ylabel(r'Reconstructed Energy [MeV]',fontsize=22)
ax2.set_xlabel(r'Neutrino Energy [MeV]',fontsize=22)
ax2.tick_params(labelsize=22)
# ax2.legend(loc="upper left", fontsize=18)

# ax2.tick_params(axis='y',which='both',labelsize=0)
ax1.set_yticks(np.arange(0,100,10))
ax1.set_xticks(np.arange(0,100,10))
ax2.set_xticks(np.arange(0,100,10))

ax1.set_xticks(np.arange(0,100,2),minor=True)
ax2.set_xticks(np.arange(0,100,2),minor=True)
# ax1.tick_params(axis='y',labelsize=15,length=7,width=2)

ax1.set_xlim(0,60)
ax2.set_xlim(0,60)
ax1.set_ylim(0,60)
ax2.set_ylim(0,60)

plt.tight_layout()
# plt.savefig("plots/QPIX_2D_Energy_Reco.png",dpi=400,bbox_inches='tight')
# plt.savefig("plots/QPIX_2D_Energy_Reco.pdf",bbox_inches='tight')
plt.show()
```
